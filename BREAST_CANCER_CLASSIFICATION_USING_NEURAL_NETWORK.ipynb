**IMPORTING DEPENDIES**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
```

**LOADING DATASETS AND PREPROCESSING**

```python
breast_cancer = load_breast_cancer()
```

```python
print(breast_cancer)
```

```python
data= pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
```

```python
#adding target column to the dataset
data['labels']= breast_cancer.target
```

**SEPERATING FEATURES AND TARGET**

```python
x= data.drop(columns='labels', axis=1)
y= data['labels']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train_std = scaler.transform(x_train)
x_test_std = scaler.transform(x_test)
```

**BUILDING NEURAL NETWORK**

```python
#importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#setting up the layers of neural networks
model =keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])

```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# training the neural network
history=model.fit(x_train_std, y_train, epochs=5,validation_split=0.1)
```

**VISUALIZATION**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc='upper right')
```

**ACCURACY OF THE MODEL ON TEST DATA**

```python
loss, accuracy = model.evaluate(x_test_std, y_test)
print(accuracy)
```
